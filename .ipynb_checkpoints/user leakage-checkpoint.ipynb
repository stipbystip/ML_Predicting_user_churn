```python
cat_feat = ["Sex", "IsSeniorCitizen", "HasPartner", "HasChild", "HasPhoneService", "HasMultiplePhoneNumbers", "HasInternetService", "HasOnlineSecurityService", "HasOnlineBackup", "HasDeviceProtection", "HasTechSupportAccess", "HasOnlineTV", "HasMovieSubscription", "HasContractPhone", "IsBillingPaperless", "PaymentMethod"]
num_feat = ['ClientPeriod', 'MonthlySpending', 'TotalSpent']
target = ["Churn"]

# Приведём TotalSpent к типу даныых float, Заменив один пропуск вида ' ' на среднее значения, так как данные
# Имеют равномерное распределение, что будет видно из графиков ни
testdf['TotalSpent'] = testdf['TotalSpent'].replace(' ',  0).astype(float)
traindf['TotalSpent'] = traindf['TotalSpent'].replace(' ', 0).astype(float)

testdf['TotalSpent'] = testdf['TotalSpent'].replace(0, testdf['TotalSpent'].mean())
traindf['TotalSpent'] = traindf['TotalSpent'].replace(0, traindf['TotalSpent'].mean())
```

# Анализ данных

```python
fig, axs = plt.subplots(8, 2)
fig.set_figheight(30)
fig.set_figwidth(10)
for i in range(16):
    row = i // 2
    col = i % 2
    train_cat_val_cnt = traindf[cat_feat[i]].value_counts()
    test_cat_val_cnt = testdf[cat_feat[i]].value_counts()
    X_axis = np.arange(len(traindf[cat_feat[i]].value_counts().index.to_list()))
    
    axs[row][col].bar(X_axis - 0.15, train_cat_val_cnt.to_list(), width=0.3, label = 'train', color='cornflowerblue')
    axs[row][col].bar(X_axis + 0.15, test_cat_val_cnt.to_list(), width=0.3, label = 'test', color='peachpuff')
    axs[row][col].set_title(cat_feat[i])
    axs[row][col].xaxis.set_major_locator(ticker.FixedLocator(X_axis))
    axs[row][col].xaxis.set_major_formatter(ticker.FixedFormatter(train_cat_val_cnt.index.to_list()))
    axs[row][col].legend()

plt.show()
    
```

```python
fig, axs = plt.subplots(2, 3)

for i in range(3):   
    axs[0][i].hist(traindf[num_feat[i]], color='limegreen')
    axs[0][i].set_title(num_feat[i])

for i in range(3):   
    axs[1][i].hist(testdf[num_feat[i]], color='mediumslateblue')
    axs[1][i].set_title(num_feat[i])

plt.show()
```

```python
fig, ax = plt.subplots()

ax.bar([str(i) for i in traindf[target[0]].value_counts().index.to_numpy()], traindf[target[0]].value_counts().to_list(), color='lightseagreen')
plt.show()
```

Видно, что количество элементов класса "0" сильно превосходит количество значений класса "1". Значит, классы не сбалансированны!

# Линейные модели

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import StandardScaler
```

```python
standard = StandardScaler()

linear_traindf = pd.concat([traindf[num_feat], pd.get_dummies(traindf[cat_feat])], axis=1)
linear_testdf = pd.concat([testdf[num_feat], pd.get_dummies(testdf[cat_feat])], axis=1)

linear_traindf = pd.DataFrame(standard.fit_transform(linear_traindf), columns=standard.get_feature_names_out())
linear_testdf = pd.DataFrame(standard.fit_transform(linear_testdf), columns=standard.get_feature_names_out())
```

```python
lr = LogisticRegression()
parameters = {'C':[i for i in np.linspace(0, 100, 100)]}

grid_search = GridSearchCV(lr,
                           parameters,
                           scoring='roc_auc',
                           n_jobs=-1,
                           cv=7)

X_train = linear_traindf.values
y_train = traindf[target].values
X_test = linear_testdf.values
```

```python
grid_search.fit(X_train, y_train.ravel())
```

```python

print('Лучший гиперпараметр: ', grid_search.best_params_['C'], '\n',
      'ROC-AUC лучшей модели: ', grid_search.best_score_)
```

```python
lr = LogisticRegression(C=grid_search.best_params_['C'])

lr.fit(X_train, y_train.ravel())
y_test = lr.predict_proba(X_test)
y_test = y_test[:, 1]
```

# Древестная модель

```python
from catboost import CatBoostClassifier
```

```python
X_train = traindf[num_feat + cat_feat].values
y_train = traindf[target].values
X_test = testdf[num_feat + cat_feat].values

cat_feat_indxs = [i for i in range(3, 18 + 1)]
```

```python
gboost = CatBoostClassifier(cat_features=cat_feat_indxs,
                            eval_metric='AUC',
                            early_stopping_rounds=10)

params_grid = {
          "iterations": [200, 150],
          "depth": [12, 15],
          "learning_rate": [0.01, 0.03, 0.05],
          "l2_leaf_reg": [5, 9]        
}

gboost.grid_search(params_grid, X=X_train, y=y_train, plot=True)
```

Лучший результат: 0.868869

```python
gboost.fit(X=X_train, y=y_train)
y_test_tree = gboost.predict_proba(X_test)
y_test_tree = y_test_tree[:, 1]
```

## Чтобы увеличить результаты моделей, в качестве ответа используем среднее значение

```python
submit = pd.DataFrame([[i, (y_test[i] + y_test_tree[i])/2] for i in range(len(y_test))], columns=['Id', 'Churn'])
submit.to_csv('submit.csv')
```
