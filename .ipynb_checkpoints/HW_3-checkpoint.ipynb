```python
def dataset_info(dataset):
    '''
Функция принимает датасет и выводит первичные данные о нем:
- 10 случайных записей
- общую информацию о датасете
- количество дубликатов
- количество пропусков
- полное описание значений датасета
    '''
    display(dataset.sample(10))
    dataset.info()
    dupl_qty_ = dataset.duplicated().sum()
    display(Markdown('**Дубликатов** - %s' % dupl_qty_))
    if dupl_qty_ > 0:
        display(dataset[dataset.duplicated()])
    display(Markdown('**Пропусков:** '))
    display(dataset.isnull().sum())
    display(dataset.describe(include='all'))
```

```python
def feature_observe(feature):
    '''
Функция принимает признак датасета и выводит информацию о нем:
- Тип признака
- Количество уникальных значений
- Список самих уникальных значений
- Распределение этих значений в искомом признаке во всех записях датасета
    '''
    _ = feature.unique()
    display(Markdown('**Признак** - %s' %feature.name))
    display(Markdown('**Тип** - %s ' %feature.dtype))
    display(Markdown('**Количество уникальных значений** - %s' %len(_)))
    display(Markdown('**Значения**:'))
    display(sorted(_.astype(str)))
    display(Markdown('**Распределение значений**:'))
    display(feature.value_counts())
    display(Markdown('**Всего записей** - %s' %len(feature)))
    display(Markdown('**Пропусков** - %s' %feature.isna().sum()))
```

```python
# Смотрим данные всех признаков:
for col in tqdm(data.columns):
    feature_observe(data[col])
    display(Markdown('---'))
```

В числовом признаке TotalSpent присутствуют записи со строковым значением - " "  
Нужно перепроверить эти записи.

```python
data['TotalSpent'] = pd.to_numeric(data['TotalSpent'], errors='coerce')
display('Пропусков - %s' %len(data[data['TotalSpent'].isna()]))
data[data['TotalSpent'].isna()]
```

Похоже, что эти клиенты только что заключили контракт (ClientPeriod = 0), у них есть стоимость ежемесячного платежа, но еще нет значения в признаке TotalSpent.
В признаке TotalSpend значение скорее всего должно появиться после окончания расчетного периода.
Нужно понимать, что такие же ситуации могут возникнуть и далее в процессе использования моделей, поэтому можно пойти двумя путями:
1. Для наших расчетов сейчас заполнить пропуски в признаке TotalSpend значением  MonthlySpend, поскольку это отражает физическую реальность на момент наступления расчетного периода
2. Заполнить пропуски " " в TotalSpent нулями, так как это отражает реальность на момент заключения нового договора и дает модели более точную информацию о состоянии новых клиентов с верным текущим статусом.

В любом случае - выбор нужно зафиксировать и делать это заполнение для всех датасетов перед запуском модели.  
Пойдем вторым путем.


```python
# Заполняем пропуски нулями:
# Сохраним индексы для последующей проверки
empty_TS_index = data[data['TotalSpent'].isna()].index

# Заполняем пропуски нулями
data.TotalSpent = data.TotalSpent.fillna(0)

# Проверяем:
data.loc[empty_TS_index]
```

```python
# Для корректного подсчета количества записей переиндексируем данные для получения уникальных значений - аналога персонального номера клиента
try:
    data = data.set_index('ID')
    display('данные уже были переиндексованы')
    display(data.sample(5))
except:
    try:
        data = data.reset_index()
        data = data.rename(columns={'index': 'ID'})
        display(data.sample(5))
    except:
        display('Возможно возникла ошибка')

# Ушедшие-потерянные клиенты
lost_cust_df = data.loc[data['Churn'] == 1]

# Оставшиеся клиенты
remain_cust_df = data.loc[data['Churn'] == 0]
```

```python
def boxplot_net(num_cols_list):
    '''
    Функция строит графики распределения количества клиентов по значениям списка числовых признаков num_cols_list
    '''
    grapf_qty=len(num_cols_list)
    plt.figure(figsize=(14,7)).suptitle('Распределение количества клиентов по значениям числовых признаков', fontsize = 14, x = 0.5)
    iter_ = 1
    for num_col in num_cols_list:

        plt.subplot(1, grapf_qty, iter_)
        data.boxplot(column=num_col)
        #plt.xlabel(num_col, fontsize=14)
        iter_+=1

    plt.tight_layout()
    plt.show()
```

```python
def cat_features_graph_bin(feature, ylabel=''):
    '''
    Функция строит графики распределения количества клиентов по значениям рассматриваемого категориального признака
    Слева - круговой график для оставшихся клиентов
    Справа - круговой график для потерянных клиентов
    '''

    # данные для признака по оставшимся клиентам
    remain_cust=remain_cust_df.groupby(feature).agg({'ID':'count'})
#    remain_cust=remain_cust_df.groupby(feature).value_counts()
    labels_1=list(remain_cust.index)

    # данные для признака по ушедшим клиентам
    lost_cust=lost_cust_df.groupby(feature).agg({'ID':'count'})
#    lost_cust=lost_cust_df.groupby(feature).value_counts()
    labels_2=list(lost_cust.index)


    plt.figure(figsize=(14,7)).suptitle('Распределение количества клиентов по значениям признака - '+feature,
                                        fontsize = 20, x = 0.5)

    # Левый график
    plt.subplot(1, 2, 1)
    plt.pie(remain_cust.values.flatten().tolist(),
            labels=labels_1,
            autopct='%1.1f%%')
    plt.xlabel('Оставшиеся клиенты', fontsize=14)

    # Правый график
    plt.subplot(1, 2, 2)
    plt.pie(lost_cust.values.flatten().tolist(),
            labels=labels_2,
            autopct='%1.1f%%')
    plt.xlabel('Ушедшие клиенты', fontsize=14)

    plt.tight_layout()
    plt.show()
```

```python
# Посмотрим на распределения значений числовых признаков
boxplot_net(num_cols)
```

# Применение линейных моделей

```python
from sklearn.linear_model import LogisticRegression, LogisticRegressionCV
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, RobustScaler, LabelEncoder, OneHotEncoder, OrdinalEncoder
from sklearn.pipeline import make_pipeline, Pipeline
from sklearn.compose import ColumnTransformer

from sklearn.metrics import roc_auc_score
```

```python
# Делим на выборки с учетом несбалансированности целевого признака (используем параметр stratify для целевого признака):

X_train, X_valid, y_train, y_valid = train_test_split(data.drop(columns=['ID','Churn'], axis=1),
                                                      data['Churn'],
                                                      test_size=0.25,
                                                      random_state=RANDOM_STATE,
                                                      stratify=data['Churn'],
                                                      shuffle=True)

# Проверка
print('Обучающая выборка: ', X_train.shape, y_train.shape)
print('Тестовая выборка: ', X_valid.shape, y_valid.shape)
```

```python
# Будем сохранять результаты обучения моделей в словаре:
cv_results = {}
```

```python
def model_score(model, preprocessor, params, cv=5, score='roc_auc'):
    '''
    Функция принимает параметры модели, перебирает параметры кроссвалидацией,
    обучает модель на тренировочной выборке.
    Возвращает значение заданной метрики (по умолчанию метрика - roc_auc),
    время обучения на обучающей выборке,
    и параметры лучшей модели
    '''
    print('Working..')
    pipeline = Pipeline(steps=[('preprocessor', preprocessor),
                               ('clf', model)]
                        )
    cv_grid = GridSearchCV(pipeline,
                           params,
                           cv=cv,
                           scoring=score,
                           n_jobs=-1)
    cv_grid.fit(X_train, y_train)

    print('Done')
    print('ROC-AUC', cv_grid.best_score_)
    print('Best params', cv_grid.best_params_)

    return {'ROC-AUC': cv_grid.best_score_,
            'Best_params' : cv_grid.best_params_,
            'Best_model' : cv_grid.best_estimator_
           }
```

```python
# Для логистической регрессии нормируем числовые признаки и кодируем категориальные признаки с помощью OHE
preprocessor_LR = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), num_cols),
        ('cat', OneHotEncoder(), cat_cols)
    ])
```

```python
# Логистическая регрессия
model = LogisticRegression(random_state=RANDOM_STATE, max_iter=500)
params = {'clf__C': [100, 10, 1, 0.1, 0.01, 0.001]
          }

cv_results[model.__class__.__name__] = model_score(model, preprocessor_LR, params)
```

Лучшее значение метрики ROC-AUC для логистической регрессии - 0.8429028267469857  
С гиперпараметром С=10


## Применение градиентного бустинга

```python
!pip install catboost
```

```python
import catboost
from catboost import CatBoostClassifier
```

```python
# Для CatBoostClassifier
# используем RobustScaler для масштабирования со сглаживанием выбросов числовых признаков
# и кодируем категориальные признаки с помощью OrdinalEncoder
preprocessor_CB = ColumnTransformer(
    transformers=[('num', RobustScaler(), num_cols),
                  ('cat', OrdinalEncoder(), cat_cols)
                  ]
    )
```

```python
# CatBoostClassifier
model = CatBoostClassifier(random_state=42, silent=True)
params = {'clf__depth': [2, 4, 6],
          'clf__iterations': [500],
          'clf__learning_rate': [0.01, 0.1, 0.5]
          }

cv_results[model.__class__.__name__] = model_score(model, preprocessor_CB, params)
```

```python
# Сравним результаты использованных моделей:
display(pd.DataFrame(cv_results).loc[:'ROC-AUC',].T)
```

Лучший результат запрошенной метрики на представленной выборке показывает модель CatBoostClassifier
ROC-AUC = 0.846345  
с гиперпараметрами 'clf__depth': 4, 'clf__iterations': 500, 'clf__learning_rate': 0.01

```python
# Заберем данные лучшей модели
best_model = cv_results['CatBoostClassifier']['Best_model']
```

```python
# best_model
```

```python
# Проверим лучшую модель на валидационной выборке
predict_proba = best_model.predict_proba(X_valid)[:, 1]
print('Метрика ROC-AUC лучшей модели:', roc_auc_score(y_valid, predict_proba))
```
