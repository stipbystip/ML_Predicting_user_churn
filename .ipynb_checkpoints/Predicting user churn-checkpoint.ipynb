```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import preprocessing
from sklearn.linear_model import LogisticRegression, LogisticRegressionCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_predict
from tqdm.notebook import tqdm
from sklearn.metrics import roc_auc_score
```

```python
X_Train_origin = pd.read_csv("train.csv")
X_Test_origin = pd.read_csv("test.csv")
sub = pd.read_csv("submission.csv")
X_Train_origin.head(10)
```

```python
print(X_Test_origin.info())
X_Train_origin.info()
```

```python
# Числовые признаки
num_cols = [
    'ClientPeriod',
    'MonthlySpending',
    'TotalSpent'
]

# Категориальные признаки
cat_cols = [
    'Sex',
    'IsSeniorCitizen',
    'HasPartner',
    'HasChild',
    'HasPhoneService',
    'HasMultiplePhoneNumbers',
    'HasInternetService',
    'HasOnlineSecurityService',
    'HasOnlineBackup',
    'HasDeviceProtection',
    'HasTechSupportAccess',
    'HasOnlineTV',
    'HasMovieSubscription',
    'HasContractPhone',
    'IsBillingPaperless',
    'PaymentMethod'
]
X_Train_origin.shape, X_Test_origin.shape
```

```python
#TotalSpent содержит пропуски
count = 0
for i in X_Train_origin['TotalSpent']:
    if i == ' ' :
        count += 1
#Заменим пропуски на NAN и изменяем тип колонки
X_Train_origin = X_Train_origin.replace(r' ', np.nan, regex= True)
X_Train_origin = X_Train_origin.astype({"TotalSpent": "float64"})
#Среднее занчения по колонке TotalSpent
mean_total_spent = X_Train_origin['TotalSpent'].mean()
#Изменяем Nan на среднее значение
X_Train_origin = X_Train_origin.replace(np.nan,mean_total_spent, regex= True)
#Тоже самое для X_Test_origin
count = 0
for i in X_Test_origin['TotalSpent']:
    if i == ' ':
        count += 1
X_Test_origin = X_Test_origin.replace(r' ', np.nan, regex = True)
X_Test_origin = X_Test_origin.astype({"TotalSpent": "float64"})
mean_total_spent = X_Test_origin['TotalSpent'].mean()
X_Test_origin = X_Test_origin.replace(np.nan, mean_total_spent, regex = True)
```

```python
dist_cat_cols = list()
for cat_col in cat_cols:
    p = X_Train_origin[cat_col].value_counts()
    dist_cat_cols.append(p)
```

```python
# Распределение целевой переменной по классам значений
plt.figure(figsize=(20, 7))
plt.hist(X_Train_origin[X_Train_origin.columns[-1]].values)
#По грифику делаем вывод, что распределения классов некритично отличается
```

```python
#Индексы числовых и категориальных переменных
u = X_Train_origin.columns
num_ind = np.array([i for i in range(len(u)) if u[i] in num_cols])
cat_ind = np.array([i for i in range(len(u)) if u[i] in cat_cols])
print(num_ind)
print(cat_ind)
```

```python
#Применим one-hot кодирование к данным
num_data = X_Train_origin[X_Train_origin.columns[num_ind]]
cat_data = X_Train_origin[X_Train_origin.columns[cat_ind]]
cat_dummies = pd.get_dummies(cat_data)

num_data2 = X_Test_origin[X_Test_origin.columns[num_ind]]
cat_data2 = X_Test_origin[X_Test_origin.columns[cat_ind]]
cat_dummies2 = pd.get_dummies(cat_data2)
```

```python
X_Train = pd.concat([num_data, cat_dummies], axis = 1)
X_Test = pd.concat([num_data2, cat_dummies2], axis = 1)
```

```python
y_train = X_Train_origin[X_Train_origin.columns[-1]].values
y_train
```

```python
X_Train.info()
```

```python
#Нормирование числовых признаков 
x1 = X_Train.iloc[:,0:3]
X_Train.iloc[:,0:3] = ((x1 - x1.min()) / (x1.max() - x1.min()))


x2 = X_Test.iloc[:,0:3]
X_Test.iloc[:,0:3] = ((x2 - x2.min()) / (x2.max() - x2.min()))
print(X_Test.shape)
X_Test.head(3)
```

```python
#Разделения на train и valid 
X_Train_split, X_Valid_split, y_train_split, y_valid_split = train_test_split(
    X_Train,  y_train, test_size = 0.2, random_state=42)
X_Train_split.shape, X_Valid_split.shape, y_train_split.shape, y_valid_split.shape
```

```python
#Решил попробовать реализовть стекинг моделей
def compute_meta_feature(model, X_Train, X_Test, y_train, cv):
    try:
        try_ans = cross_val_predict(model, X_Train, y_train, cv=cv, method='predict_proba')[:, 1]
        model.fit(X_Train, y_train)
        return try_ans, model.predict_proba(X_Test)[:, 1]
    except Exception:
        try_ans = cross_val_predict(model, X_Train, y_train, cv=cv, method='predict_proba')[:, 1]
        model.fit(X_Train, y_train)
        return try_ans, model.predict(X_Test)[:, 1]
```

```python
stack_models = []
stack_models.append(LogisticRegression(max_iter = 1000))
stack_models.append(RandomForestClassifier(max_depth=3, n_estimators=50, n_jobs=-1))
stack_models.append(RandomForestClassifier(max_depth=7, n_estimators=50, n_jobs=-1))
stack_models.append(RandomForestClassifier(max_depth=7, n_estimators=50, n_jobs=-1))
stack_models.append(DecisionTreeClassifier(max_depth=4))
```

```python
meta_features_train = np.zeros((X_Train.shape[0], 0))
meta_features_test = np.zeros((X_Test.shape[0], 0))
meta_features_test.shape
```

```python
for model in tqdm(stack_models):
    train, test = compute_meta_feature(model, X_Train, X_Test, y_train, 5)
    meta_features_train = np.append(meta_features_train, train.reshape(train.size, 1), axis=1)
    meta_features_test = np.append(meta_features_test, test.reshape(test.size, 1), axis=1)
```

```python
stacking_model = LogisticRegression()
stacking_model.fit(meta_features_train, y_train)
```

```python
sub['Churn'] = stacking_model.predict_proba(meta_features_test)[:, 1]
sub.to_csv('./my_submission.csv',  index=False)
```
