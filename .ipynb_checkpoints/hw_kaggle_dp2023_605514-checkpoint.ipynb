# Метрика

Перед решением любой задачи важно понимать, как будет оцениваться ваше решение. В данном случае мы используем стандартную для задачи классификации метрику ROC-AUC. Ее можно вычислить, используя только предсказанные вероятности и истинные классы без конкретного порога классификации + она раотает даже если классы в данных сильно несбалансированны (примеров одного класса в десятки раз больше примеров длугого). Именно поэтому она очень удобна для соревнований.

Посчитать ее легко:


```python
from sklearn.metrics import roc_auc_score

y_true = [
    0,
    1,
    1,
    0,
    1
]

y_predictions = [
    0.1,
    0.9,
    0.4,
    0.6,
    0.61
]

roc_auc_score(y_true, y_predictions)
```

## Применение линейных моделей (3 балла)

1) Обработайте данные для того, чтобы к ним можно было применить LogisticRegression. Т.е. отнормируйте числовые признаки, а категориальные закодируйте с помощью one-hot-encoding'а.

2) С помощью кроссвалидации или разделения на train/valid выборку протестируйте разные значения гиперпараметра C и выберите лучший (можно тестировать С=100, 10, 1, 0.1, 0.01, 0.001) по метрике ROC-AUC.

Если вы разделяете на train/valid, то используйте LogisticRegressionCV. Он сам при вызове .fit() подберет параметр С. (не забудьте передать scroing='roc_auc', чтобы при кроссвалидации сравнивались значения этой метрики, и refit=True, чтобы при потом модель обучилась на всем датасете с лучшим параметром C).


(более сложный вариант) Если вы будете использовать кроссвалидацию, то преобразования данных и LogisticRegression нужно соединить в один Pipeline с помощью make_pipeline, как это делалось во втором семинаре. Потом pipeline надо передать в GridSearchCV. Для one-hot-encoding'a можно испльзовать комбинацию LabelEncoder + OneHotEncoder (сначала превращаем строчки в числа, а потом числа првращаем в one-hot вектора.)

```python
from sklearn.linear_model import LogisticRegression, LogisticRegressionCV
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, RobustScaler, LabelEncoder, OneHotEncoder
from sklearn.pipeline import make_pipeline, Pipeline
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
```

Выпишите какое лучшее качество и с какими параметрами вам удалось получить

```python
# пайплайн для числовых колонок
num_cols_pipline = Pipeline([
    #("le", LabelEncoder()),
    ('imputer', SimpleImputer(strategy='mean')), # заменить пустые числовые значения на среднее
    ('scale', StandardScaler()) # нормировка числовых колонок
])
# пайплайн для категориальных колонок
cat_cols_pipeline = Pipeline([
    ('ohe', OneHotEncoder(drop='first')) # преобразовать категориальные колонки
])
# общий пайплайн колонок
column_transformer = ColumnTransformer([
    ('num_cols', num_cols_pipline, num_cols),
    ('cat_cols', cat_cols_pipeline, cat_cols),
])
```

```python
X_train = data.iloc[:, :-1] # выделить нужные колонки без целевой колонки
y_train = data.iloc[:, -1] # выделеить целевую колонку
```

```python
X_train.head()
```

```python
y_train.head()
```

```python
pd.DataFrame(column_transformer.fit_transform(X_train))
```

**LogisticRegression**

```python
from sklearn.linear_model import LogisticRegression
```

```python
logreg_pipeline = make_pipeline(column_transformer, LogisticRegression())
logreg_params = {
    'logisticregression__solver': ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga'],
    'logisticregression__penalty': ['l1', 'l2', 'elasticnet'],
    'logisticregression__max_iter': [100, 300, 500],
    'logisticregression__C': [100, 10, 1, 0.1, 0.01, 0.001]
}
logreg_grid = GridSearchCV(estimator=logreg_pipeline, param_grid=logreg_params, n_jobs=-1, cv=5, scoring='roc_auc')
```

```python
logreg_grid.fit(X_train, y_train)
```

```python
print('Best params:', logreg_grid.best_params_)
print('Best score:', logreg_grid.best_score_)
```

```python
knc_grid.fit(X_train, y_train)
```

Мы будем использовать catboost, поэтому нам не надо кодировать категориальные признаки. Catboost сделает это сам (в .fit() надо передать cat_features=cat_cols). А численные признаки нормировать для моделей, основанных на деревьях не нужно.

1) Разделите выборку на train/valid. Протестируйте catboost cо стандартными параметрами.

```python
!pip install catboost
```

```python
from catboost import CatBoostClassifier
```

```python
X_train_cb, X_valid_cb, y_train_cb, y_valid_cb = train_test_split(X_train, y_train, test_size=.2, random_state=42)
```

**CatBoost. Стандартными параметры**

```python
cbc_model = CatBoostClassifier(cat_features=cat_cols, n_estimators=200, verbose=False)
```

```python
cbc_model.fit(X_train_cb, y_train_cb)
```

```python
roc_auc_score(y_valid_cb, cbc_model.predict_proba(X_valid_cb)[:, -1])
```

**CatBoost. Поиск оптимальной модели**

Используется random search.

```python
gcbc_model = CatBoostClassifier(
    loss_function='Logloss',
    silent=True,
    cat_features=cat_cols,
    eval_metric='AUC'
)
```

```python
gcbc_model.randomized_search(
    {
        'learning_rate': [0.03, 0.05, 0.07, 0.08, 0.1],
        'depth': [4, 5, 6, 7, 8, 9, 10],
        'l2_leaf_reg': [1, 3, 5, 7, 9],
        'rsm': np.flip(np.linspace(0.02, 1, 5)),
    },
    X=X_train,
    y=y_train,
    plot=True
)
```

```python
roc_auc_score(y_train, gcbc_model.predict_proba(X_train)[:, -1])
```

Выпишите, какое лучшее качество и с какими параметрами вам удалось получить

Наилучшее качество показал CatBoostClassifier с использованием random search.
